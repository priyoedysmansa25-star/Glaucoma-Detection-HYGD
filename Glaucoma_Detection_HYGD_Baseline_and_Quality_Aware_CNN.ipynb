# 11. Membangun Arsitektur Baseline CNN

```python
model = models.Sequential([

    layers.Conv2D(
        32,
        (3,3),
        activation='relu',
        input_shape=(224,224,3)
    ),

    layers.MaxPooling2D(),

    layers.Conv2D(
        64,
        (3,3),
        activation='relu'
    ),

    layers.MaxPooling2D(),

    layers.Conv2D(
        128,
        (3,3),
        activation='relu'
    ),

    layers.MaxPooling2D(),

    layers.Flatten(),

    layers.Dense(
        128,
        activation='relu'
    ),

    layers.Dropout(0.5),

    layers.Dense(
        1,
        activation='sigmoid'
    )
])

model.summary()
```

# 12. Kompilasi Model Baseline CNN

```python
model.compile(
optimizer='adam',
loss='binary_crossentropy',
metrics=['accuracy']
)
```

```python
correct_image_folder = '/content/hillel-yaffe-glaucoma-dataset-hygd-a-gold-standard-annotated-fundus-dataset-for-glaucoma-detection-1.0.0/Images/'

df['filepath'] = correct_image_folder + df['Image Name']


gss = GroupShuffleSplit(
    n_splits=1,
    test_size=0.3,
    random_state=42
)

train_idx, temp_idx = next(
    gss.split(
        df,
        groups=df['Patient']
    )
)

train_df = df.iloc[train_idx]
temp_df = df.iloc[temp_idx]

gss_val_test = GroupShuffleSplit(
    n_splits=1,
    test_size=0.5, # 50% for test, 50% for validation from temp_df
    random_state=42
)

val_idx, test_idx = next(
    gss_val_test.split(
        temp_df,
        groups=temp_df['Patient']
    )
)

val_df = temp_df.iloc[val_idx]
test_df = temp_df.iloc[test_idx]


train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    horizontal_flip=True
)

val_gen = ImageDataGenerator(
    rescale=1./255
)


train_generator = train_gen.flow_from_dataframe(
    train_df,
    x_col='filepath',
    y_col='Label',
    class_mode='binary',
    target_size=(224,224),
    batch_size=32
)

val_generator = val_gen.flow_from_dataframe(
    val_df,
    x_col='filepath',
    y_col='Label',
    class_mode='binary',
    target_size=(224,224),
    batch_size=32
)

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10
)
```

# 13. Grafik Akurasi Model

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['Train','Validation'])
plt.show()
```

# 14. EVALUASI BASELINE CNN

```python
# Create test_generator
test_generator = val_gen.flow_from_dataframe(
    test_df,
    x_col='filepath',
    y_col='Label',
    class_mode='binary',
    target_size=(224, 224),
    batch_size=32,
    shuffle=False
)

pred_prob = model.predict(test_generator)

pred = (pred_prob > 0.5).astype(int)

y_true = test_generator.classes
```

# 15. METRIK BASELINE CNN

```python
print('Accuracy:', accuracy_score(y_true, pred))
print('Precision:', precision_score(y_true, pred))
print('Recall:', recall_score(y_true, pred))
print('F1-score:', f1_score(y_true, pred))
print('ROC AUC:', roc_auc_score(y_true, pred_prob))
```

# 16. CONFUSION MATRIX BASELINE CNN

```python
cm = confusion_matrix(y_true, pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

# BAGIAN 2 — QUALITY-AWARE CNN

# 18. Seleksi Data Berdasarkan Quality Score

```python
quality_df = df[df['Quality Score'] > 3]

print(len(df))
print(len(quality_df))
```

# 19. Pembagian Data Quality-Aware

```python
gss = GroupShuffleSplit(
    n_splits=1,
    test_size=0.3,
    random_state=42
)

train_idx_q, temp_idx_q = next(
    gss.split(
        quality_df,
        groups=quality_df['Patient']
    )
)

train_df_q = quality_df.iloc[train_idx_q]
temp_df_q = quality_df.iloc[temp_idx_q]
```

# 20. VALIDATION DAN TEST QUALITY-AWARE

```python
gss2 = GroupShuffleSplit(
    n_splits=1,
    test_size=0.5,
    random_state=42
)

val_idx_q, test_idx_q = next(
    gss2.split(
        temp_df_q,
        groups=temp_df_q['Patient']
    )
)

val_df_q = temp_df_q.iloc[val_idx_q]
test_df_q = temp_df_q.iloc[test_idx_q]
```

# 21. GENERATOR QUALITY-AWARE

```python
train_gen_q = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    horizontal_flip=True
)

val_gen_q = ImageDataGenerator(
    rescale=1./255
)
```

# 22. Membuat Generator Data Quality-Aware

```python
train_generator_q = train_gen_q.flow_from_dataframe(
    train_df_q,
    x_col='filepath',
    y_col='Label',
    class_mode='binary',
    target_size=(224,224),
    batch_size=32
)

val_generator_q = val_gen_q.flow_from_dataframe(
    val_df_q,
    x_col='filepath',
    y_col='Label',
    class_mode='binary',
    target_size=(224,224),
    batch_size=32
)

test_generator_q = val_gen_q.flow_from_dataframe(
    test_df_q,
    x_col='filepath',
    y_col='Label',
    class_mode='binary',
    target_size=(224,224),
    batch_size=32,
    shuffle=False
)
```

# 23. MEMBANGUN QUALITY-AWARE CNN

```python
model_q = models.Sequential([

layers.Conv2D(
32,
(3,3),
activation='relu',
input_shape=(224,224,3)
),

layers.MaxPooling2D(),

layers.Conv2D(
64,
(3,3),
activation='relu'
),

layers.MaxPooling2D(),

layers.Conv2D(
128,
(3,3),
activation='relu'
),

layers.MaxPooling2D(),

layers.Flatten(),

layers.Dense(
128,
activation='relu'
),

layers.Dropout(0.5),

layers.Dense(
1,
activation='sigmoid'
)

])
```

# 24. KOMPILASI QUALITY-AWARE CNN

```python
model_q.compile(
optimizer='adam',
loss='binary_crossentropy',
metrics=['accuracy']
)
```

# 25. TRAINING QUALITY-AWARE CNN

```python
history_q = model_q.fit(
train_generator_q,
validation_data=val_generator_q,
epochs=10
)
```

# 26. EVALUASI QUALITY-AWARE CNN

```python
pred_prob_q = model_q.predict(test_generator_q)

pred_q = (pred_prob_q > 0.5).astype(int)

y_true_q = test_generator_q.classes
```

# 27. METRIK QUALITY-AWARE CNN

```python
print('accuary:',accuracy_score(y_true_q,pred_q))

print('precision:',precision_score(y_true_q,pred_q))

print('recall:',recall_score(y_true_q,pred_q))

print('f1-score:',f1_score(y_true_q,pred_q))

print('roc-auc:',roc_auc_score(y_true_q,pred_prob_q))
```

# 28. Confusion Matrix Quality-Aware CNN

```python
cm_q = confusion_matrix(
y_true_q,
pred_q
)

sns.heatmap(
cm_q,
annot=True,
fmt='d'
)

plt.show()
```
